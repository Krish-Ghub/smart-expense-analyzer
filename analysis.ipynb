```python
import pandas as pd

data = [
    {"id": 3, "amount": 50.0, "category": "Food", "date": "2026-04-04"}, 
    {"id": 4, "amount": 222.0, "category": "Food", "date": "2026-04-04"}, 
    {"id": 1, "amount": 120.0, "category": "Food", "date": "2026-04-03"}, 
    {"id": 2, "amount": 120.0, "category": "Food", "date": "2026-04-03"}
]

df = pd.DataFrame(data)

print(df)

print(df["amount"].sum())


print(df.groupby("category")["amount"].sum())

df["date"] = pd.to_datetime(df["date"])
print(df.groupby("date")["amount"].sum())

food_spending = df[df["category"] == "Food"]["amount"].sum()
total = df["amount"].sum()

if food_spending > 0.4 * total:
    print("⚠️ You spend too much on food")
```

```python

import matplotlib.pyplot as plt

category_sum = df.groupby("category")["amount"].sum()

category_sum.plot(kind='bar')
plt.title("Spending by Category")
plt.show()

```

```python
df["date"] = pd.to_datetime(df["date"])

daily = df.groupby("date")["amount"].sum()

daily.plot()
plt.title("Daily Spending Trend")
plt.show()
```

```python
total = df["amount"].sum()

food = df[df["category"] == "Food"]["amount"].sum()

if food > 0.4 * total:
    print("⚠️ High spending on food")

if df["amount"].mean() > 200:
    print("⚠️ Your average expense is high")
```

```python

from sklearn.linear_model import LinearRegression
import numpy as np

df["day"] = range(len(df))

X = df[["day"]]
y = df["amount"]

model = LinearRegression()
model.fit(X, y)

prediction = model.predict([[len(df)]])
print("Next expense prediction:", prediction[0])
```
